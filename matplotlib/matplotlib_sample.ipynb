# Step 1: Import Libraries and Plot Sine (Single Plot)
In this first cell we import the libraries, generate the data, and plot the sine curve in a simple figure. (Since this cell creates a plot, it will display something.)

```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, clear_output, Image
%matplotlib notebook


# Generate data: 100 points between 0 and 10
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot sine only (a single-axes figure)
plt.figure(figsize=(6, 4))
plt.plot(x, y_sin, label='Sine')
plt.xlabel("X Axis")
plt.ylabel("Sine Value")
plt.title("Step 1: Sine Wave")
plt.legend()
plt.show()
```

# Step 2: Create a Figure with Two Subplots (Sine and Cosine Side-by-Side)
Now we create a figure with two subplots arranged in 1 row and 2 columns. In the left subplot we plot sine, and in the right subplot we plot cosine.

```python
# Create a figure with two subplots (side-by-side)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6, 3), dpi=150)

# Plot sine on the first subplot
ax1.plot(x, y_sin, label='Sine', color='blue')
ax1.set_title("Step 2: Sine Wave")
ax1.set_xlabel("X Axis")
ax1.set_ylabel("Sine Value")
ax1.legend()

# Plot cosine on the second subplot
ax2.plot(x, y_cos, label='Cosine', color='orange')
ax2.set_title("Step 2: Cosine Wave")
ax2.set_xlabel("X Axis")
ax2.set_ylabel("Cosine Value")
ax2.legend()

plt.show()

```

# Step 3: Customize the Sine Subplot (ax1)
In this cell we further customize the left subplot by setting axis limits and adding grid lines.

```python
# Step 3: Customize the sine subplot (ax1)
ax1.set_xlim(0, 10)
ax1.set_ylim(-1.5, 1.5)
ax1.grid(True)
ax1.set_title("Step 3: Sine Wave (Customized)")

clear_output(wait=True)  # Clear the previous output cell
display(fig)             # Display the updated figure


```

# Step 4: Plot the Sine Wave on the First Subplot
We plot the sine function on the first subplot (ax1) with custom styling.

```python
# Plot sine on the first subplot (ax1)
ax1.plot(
    x, y_sin,
    color="#1f77b4",          # Blue color (hex code)
    linestyle='-',            # Solid line
    linewidth=2,
    marker='o',               # Circle markers
    markersize=5,
    markerfacecolor='white',  # Hollow markers
    markeredgecolor="#1f77b4"
)

clear_output(wait=True)  # Clear the previous output cell
display(fig)             # Display the updated figure
```

# Step 5: Customize the First Subplot (Sine Plot)
We add a title, set the axis limits, and enable grid lines for the first subplot.

```python
# Customize the first subplot
ax1.set_title("Sine Wave", fontsize=14)
ax1.set_xlim(0, 10)
ax1.set_ylim(-1.5, 1.5)
ax1.grid(True)

clear_output(wait=True)  # Clear the previous output cell
display(fig)             # Display the updated figure
```

# Step 6: Plot the Cosine Wave on the Second Subplot
Now, we plot the cosine function on the second subplot (ax2) with a different style.

```python
# Plot cosine on the second subplot (ax2)
ax2.plot(
    x, y_cos,
    color="#ff7f0e",          # Orange color (hex code)
    linestyle='--',           # Dashed line
    linewidth=2,
    marker='s',               # Square markers
    markersize=5,
    markerfacecolor='white',
    markeredgecolor="#ff7f0e"
)

clear_output(wait=True)  # Clear the previous output cell
display(fig)             # Display the updated figure
```

# Step 7: Customize the Second Subplot (Cosine Plot)
We add a title, set the axis limits, and enable grid lines for the cosine plot.

```python
# Customize the second subplot
ax2.set_title("Cosine Wave", fontsize=14)
ax2.set_xlim(0, 10)
ax2.set_ylim(-1.5, 1.5)
ax2.grid(True)

clear_output(wait=True)  # Clear the previous output cell
display(fig)             # Display the updated figure
```

# Step 8: Adjust Layout 
We adjust the overall layout of the figure to ensure that titles and labels do not overlap.

```python
fig.tight_layout(rect=[0, 0.03, 1, 0.85])

display(fig)             # Display the updated figure
```

# Step 9: Save the Figure as a High-Resolution Image
Before displaying, we save the figure to a file. We specify a high DPI (300) for a publication-quality image.

```python
# Save the figure to a PNG file with high resolution (300 dpi)
fig.savefig("matplotlib_sample.png", dpi=300)

# Display the saved image directly
display(Image(filename="matplotlib_sample.png"))
```
